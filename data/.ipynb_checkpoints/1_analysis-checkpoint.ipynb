## 2. task data

```python
file_path = '1_nill/task_data_10_09_2024_19_18_13.csv'
task = pd.read_csv(file_path)
task
```

## 2.1: stimus types:

```python
balloon_type_counts = task['balloonType'].value_counts()
plt.figure(figsize=(8,6))
balloon_type_counts.plot(kind='bar')
plt.title('Count of Different Balloon Types')
plt.xlabel('Balloon Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 2.2: mean RT:

```python
reaction_time_mean = task['reactionTime (ms)'].mean()
print("Mean reaction time (ms):", reaction_time_mean)
```

### 2.2.1: stim based mean RT:

```python
reaction_time_by_balloon = task.groupby('balloonType')['reactionTime (ms)'].mean().sort_values(ascending=False)

plt.figure(figsize=(8,6))
reaction_time_by_balloon.plot(kind='bar')
plt.title('Mean Reaction Time by Balloon Type')
plt.xlabel('Balloon Type')
plt.ylabel('Mean Reaction Time (ms)')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

## 2.3: mean IT:

```python
reaction_time_mean = task['inflationTime (ms)'].mean()
print("Mean inflation time (ms):", reaction_time_mean)
```

### 2.3.1: stim based mean IT:

```python
reaction_time_by_balloon = task.groupby('balloonType')['inflationTime (ms)'].mean().sort_values(ascending=False)

plt.figure(figsize=(8,6))
reaction_time_by_balloon.plot(kind='bar')
plt.title('Mean Inflation Time by Balloon Type')
plt.xlabel('Balloon Type')
plt.ylabel('Mean Inflation Time (ms)')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

## 2.4: outcome status:
gray and special stimulus were removed!

```python
filtered_task = task[~task['balloonType'].str.contains('gray|special', case=False)]

outcome_percentage = filtered_task['outcome'].value_counts(normalize=True) * 100
plt.figure(figsize=(8,6))
outcome_percentage.plot(kind='bar')
plt.title('Outcome Percentage Distribution')
plt.xlabel('Outcome')
plt.ylabel('Percentage(%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 2.4.1: color based outcome:

```python
outcome_by_balloon = filtered_task.groupby('balloonType')['outcome'].value_counts(normalize=True).unstack() * 100
outcome_by_balloon.plot(kind='bar', figsize=(10, 6))
plt.title('Outcome Percentage by Balloon Type')
plt.xlabel('Balloon Type')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
